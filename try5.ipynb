```python
import pandas as pd
import numpy as np
import pymorphy2
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('train.csv')
data.head()
```

```python
data['word1'] = data['Sample'].str.split().str.get(0)
data['word2'] = data['Prediction'].str.split().str.get(1)
data['word'] = data['Sample'].str.split().str.get(1)
#data.head()
data
```

```python
word1 = data['word1'].value_counts().to_dict()
word2 = data['word2'].value_counts().to_dict()
tmpword2 = word2.copy()
for word in word1:
    if word in tmpword2:
        tmpword2[word] += word1[word]
tmpwords = word1.copy()
tmpwords.update(tmpword2)
words = {}
for word in tmpwords:
    if word[:2] in words:
        # добавляем слово для данных 2ух букв
        words[word[:2]].update({word : 1.0 * tmpwords[word]})
    else:
        words[word[:2]] = {word : 1.0 * tmpwords[word]}

words
#делаем словарь с частотностью слов
```

```python
tmpcollocations = data['Prediction'].value_counts().to_dict()
collocations = dict()
for col, val in tmpcollocations.items():
    
    w1, w2 = col.split()
    key = w1 +' '+ w2[:2]
    if key in collocations:
        collocations[key][w2] = val
    else:
        collocations[key]={w2: val}
collocations
#делаем словарь с частотностью словосочетаний
```
